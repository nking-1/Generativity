```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k_B = 1.380649e-23  # Boltzmann constant in J/K
hbar = 1.0545718e-34  # Reduced Planck constant in J·s
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
c = 299792458  # Speed of light in m/s

# Define the mass range for the black hole (in solar masses)
M_sun = 1.98847e30  # Mass of the Sun in kg
M_range = np.linspace(1 * M_sun, 100 * M_sun, 100)  # Masses in kg

# Schwarzschild radius, A (area), and entropy S for each M
R_s = 2 * G * M_range / c**2  # Schwarzschild radius in meters
A = 4 * np.pi * R_s**2  # Surface area of the black hole
S_BH = (k_B * A) / (4 * (G * hbar / c**3))  # Bekenstein-Hawking entropy

# Hawking radiation power and temperature
T_H = hbar * c**3 / (8 * np.pi * G * k_B * M_range)  # Hawking temperature
P_H = (hbar * c**6) / (15360 * np.pi * G**2 * M_range**2)  # Power radiated by the black hole

# Entropy change rate dS/dt
dS_dt = P_H / T_H  # Rate of entropy change

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot S vs M
ax1.set_xlabel("Black Hole Mass (kg)")
ax1.set_ylabel("Entropy S (J/K)", color='tab:blue')
ax1.plot(M_range, S_BH, color='tab:blue', label='S (Entropy)')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for dS/dt
ax2 = ax1.twinx()
ax2.set_ylabel("Entropy Change Rate dS/dt (J/K/s)", color='tab:green')
ax2.plot(M_range, dS_dt, color='tab:green', label='dS/dt')
ax2.tick_params(axis='y', labelcolor='tab:green')

# Title and show plot
plt.title("Entropy and Rate of Entropy Change for Varying Black Hole Mass")
fig.tight_layout()
plt.savefig("blackhole_entropy_vs_mass.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k_B = 1.380649e-23  # Boltzmann constant in J/K
hbar = 1.0545718e-34  # Reduced Planck constant in J·s
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
c = 299792458  # Speed of light in m/s

# Mass range (1 to 100 solar masses)
M_sun = 1.98847e30  # kg
M_range = np.linspace(1 * M_sun, 100 * M_sun, 500)

# Schwarzschild radius and area
R_s = 2 * G * M_range / c**2
A = 4 * np.pi * R_s**2

# Entropy S_BH
l_p_squared = hbar * G / c**3
S_BH = k_B * A / (4 * l_p_squared)

# Hawking temperature and power
T_H = hbar * c**3 / (8 * np.pi * G * k_B * M_range)
P_H = hbar * c**6 / (15360 * np.pi * G**2 * M_range**2)

# Entropy change rate
dS_dt = P_H / T_H

# Information flow: I_max = S * dS/dt
I_max = S_BH * dS_dt

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Entropy plot
ax1.set_xlabel("Black Hole Mass (kg)")
ax1.set_ylabel("Entropy S (J/K)", color='tab:blue')
ax1.plot(M_range, S_BH, color='tab:blue', label='S (Entropy)')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create second y-axis for I_max
ax2 = ax1.twinx()
ax2.set_ylabel("Information Flow I_max (J²/K²·s)", color='tab:red')
ax2.plot(M_range, I_max, color='tab:red', label='I_max = S · dS/dt')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add combined legend
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Title and layout
plt.title("Entropy and Information Flow vs Black Hole Mass")
fig.tight_layout()
plt.savefig("blackhole_information_flow.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11       # m^3 kg^-1 s^-2
c = 299792458         # m/s
M_sun = 1.98847e30    # kg
M = 10 * M_sun        # Black hole mass (10 solar masses)
r_s = 2 * G * M / c**2  # Schwarzschild radius

# Radial positions from 1.1 r_s to 10 r_s
r = np.linspace(1.1 * r_s, 10 * r_s, 500)

# Time dilation factor (gravitational redshift)
# This is the factor by which coordinate time slows down relative to proper time
time_dilation = np.sqrt(1 - r_s / r)

# Assume constant intrinsic complexity S
S = 1.0  # arbitrary units

# Information flow from the perspective of a distant observer
# As time dilation increases (i.e., time slows), dS/dt goes to 0 => I -> 0
I = S * time_dilation  # since dS/dt ~ 1/time_dilation

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(r / r_s, I, label=r"$\mathcal{I}(r) = S \cdot \frac{dS}{dt}$", color='tab:blue')
plt.axvline(1, color='black', linestyle='--', label='Event Horizon ($r = r_s$)')
plt.xlabel("Radial Distance $r$ (in units of Schwarzschild radius $r_s$)")
plt.ylabel("Information Flow $\mathcal{I}(r)$ (arbitrary units)")
plt.title("Information Flow of a Particle Approaching a Black Hole")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11       # m^3 kg^-1 s^-2
c = 299792458         # m/s
hbar = 1.0545718e-34  # J·s
k_B = 1.380649e-23    # J/K
M_sun = 1.98847e30    # kg

# Initial conditions
M0 = 10 * M_sun  # initial black hole mass
kappa = 1e-17  # evaporation rate scaling constant
t_total = (M0**3) / kappa  # approximate lifetime of the black hole
t = np.linspace(0, t_total, 1000)

# Split time into three phases:
# 1. Pre-fall (first 20%)
# 2. Near-horizon suppression (middle 60%)
# 3. Evaporation and information return (final 20%)

t1 = t[:200]
t2 = t[200:800]
t3 = t[800:]

# Phase 1: Particle is far from black hole, normal info flow
I1 = np.full_like(t1, 1.0)

# Phase 2: Particle nears horizon and becomes unresolved
# We'll simulate a drop in I toward a nonzero minimum
I2 = np.linspace(1.0, 0.01, len(t2))

# Phase 3: Black hole evaporates, I climbs back up
I3 = np.linspace(0.01, 0.8, len(t3))

# Combine
t_combined = np.concatenate([t1, t2, t3]) / (3600 * 24 * 365)  # convert to years
I_combined = np.concatenate([I1, I2, I3])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t_combined, I_combined, color='tab:purple', label=r"$\mathcal{I}(t)$")
plt.axvline(t_combined[200], color='gray', linestyle='--', label='Approaching Horizon')
plt.axvline(t_combined[800], color='gray', linestyle='--', label='Evaporation Phase Begins')
plt.xlabel("Time (years)")
plt.ylabel("Information Flow $\mathcal{I}(t)$ (normalized units)")
plt.title("Information Flow of a Particle: Fall, Suppression, and Reemergence")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
